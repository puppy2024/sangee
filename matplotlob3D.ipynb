```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]
z=[3,4,6,1,2,5]
ax.plot3D(x,y,z,color='red')
ax.set_title("3d")
plt.show()
```

```python
sub=['t','e','m','s','ss']
marks=[98,85,72,95,92]
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,8))

ax[0][0].plot(sub,marks,color='green')
ax[0][0].set_title("line plot")

ax[1][0].bar(sub,marks,color='red')
ax[1][0].set_title("bar plot")

ax[0][1].scatter(sub,marks,color='orange')
ax[0][1].set_title("scatter plot")

ax[1][1].pie(marks,labels=sub,autopct="%1.0f%%",startangle=90)
ax[1][1].set_title("pie plot")
ax[1][1].axis("equal")
plt.tight_layout()
plt.show()
```
