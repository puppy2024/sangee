```python
import matplotlib.pyplot as plt
```

```python
food=['fried rice','biriyani','chappati','pizza','burger']
price=['150','200','100','499','199']
```

```python
plt.plot(food,price)
plt.show()
```

```python
plt.bar(food,price)
plt.show()
```

```python
colors=['red','gray','pink','violet','orange']
plt.bar(food,price,color=colors,alpha=0.6,edgecolor='black',linewidth=2,linestyle='--',label='menu')
plt.grid(axis='y',linestyle='-.')
plt.xlabel("food")
plt.ylabel("price")
plt.title("Hotel")
plt.legend()
plt.show()
```

```python
plt.pie(price,labels=food,autopct='%1.0f%%')
plt.show()
```

```python
plt.plot(food,price,color='black',linewidth=0.8,linestyle='-.',marker='+',label="pricetag")
plt.legend()
plt.show()
```
